```python
# 统计总氮百分比异常值的行数和占比
lost_count = sum(data2_1["总氮百分比"]>1)
lost_count_prop = lost_count/len(data2_1)
# 统计P2O5百分比异常值的行数和占比
lost_P205 = sum(data2_1["P2O5百分比"]>1)
lost_P205_prop = lost_P205/len(data2_1)
# 统计K2O百分比异常值的行数和占比
lost_K2O = sum(data2_1["K2O百分比"]>1)
lost_K2O_prop = lost_K2O/len(data2_1)
# 统计有机质百分比异常值的行数和占比
lost_jz = sum(data2_1["有机质百分比"]>1)
lost_jz_prop = lost_jz/len(data2_1)
pd.DataFrame({'异常值数量':[lost_count,lost_P205,lost_K2O,lost_jz], '异常值占比':[lost_count_prop,lost_P205_prop,lost_K2O_prop,lost_jz_prop]},index=["总氮百分比","P2O5百分比","K2O百分比","有机质百分比"])
```

```python
# 统计总氮百分比异常值的行数和占比
lost_count1 = sum(data2_1["总氮百分比"]<0)
lost_count_prop1 = lost_count1/len(data2_1)
# 统计P2O5百分比异常值的行数和占比
lost_P2051 = sum(data2_1["P2O5百分比"]<0)
lost_P205_prop1 = lost_P2051/len(data2_1)
# 统计K2O百分比异常值的行数和占比
lost_K2O1 = sum(data2_1["K2O百分比"]<0)
lost_K2O_prop1 = lost_K2O1/len(data2_1)
# 统计有机质百分比异常值的行数和占比
lost_jz1 = sum(data2_1["有机质百分比"]<0)
lost_jz_prop1 = lost_jz1/len(data2_1)
pd.DataFrame({'异常值数量':[lost_count1,lost_P2051,lost_K2O1,lost_jz1], '异常值占比':[lost_count_prop1,lost_P205_prop1,lost_K2O_prop1,lost_jz_prop1]},index=["销售数量","P2O5百分比","K2O百分比","有机质百分比"])
```
