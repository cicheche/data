取值范围 R=max{X}-min{X} 由于取值范围是客观存在的，当组数确定之后，组距也就随之确定了，两者成反比例关系。设H为组数，d为组距，即有组距d=R/H。为了便于数据分组和组限的划定，组距一般取5，10的整数倍。

```python
result2_2.describe()
```

```python
# 有机
oc = (result2_2['有机质百分比'].max()-result2_2['有机质百分比'].min())/10
result2_2['有机质百分比标签'] = result2_2['有机质百分比'].map(lambda x: x/oc)
result2_2
```

```python
pd.cut(result2_2['有机质百分比标签'],10)
```

```python
bins = [-0.0008,0.08,0.17,0.26,0.35,0.44,0.53,0.62,0.71,0.80,0.9] 
num = [i for i in range(1,11)]
result2_2['有机质百分比标签'] = pd.cut(result2_2['有机质百分比'],bins,labels=num)
result2_2
```

```python
# 无机
oc = (result2_2['总无机养分百分比'].max()-result2_2['总无机养分百分比'].min())/10
result2_2['总无机养分百分比标签'] = result2_2['总无机养分百分比'].map(lambda x: x/oc)
result2_2
```

```python
pd.cut(result2_2['总无机养分百分比'],4)
```

```python
bins = [0.0482, 0.0871, 0.125, 0.163, 0.201,0.239,0.277,0.315,0.352,0.391,0.43] 
num = [i for i in range(1,11)]
result2_2['总无机养分百分比标签'] = pd.cut(result2_2['总无机养分百分比'],bins,labels=num)
result2_2
```

```python
result2_2['分组标签'] = result2_2[['总无机养分百分比标签', '有机质百分比标签']].apply(tuple, axis=1)
```
