```python
import pandas as pd
import numpy as np
guanghu_data = pd.read_excel('附件2.xlsx')
guanghu_data
```

```python
# 查重，去重
guanghu_data[guanghu_data['正式登记证号'].duplicated()]
data = guanghu_data.drop_duplicates(['正式登记证号'])
```

```python
result2_1 = guanghu_data.loc[guanghu_data['产品通用名称'] == '复混肥料']
result2_1
```

```python
# 无机
oc = (result2_1['总无机养分百分比'].max()-result2_1['总无机养分百分比'].min())/10
result2_1['总无机养分百分比标签'] = result2_1['总无机养分百分比'].map(lambda x: x/oc)
result2_1
```

```python
pd.cut(result2_1['总无机养分百分比'],10)
```

```python
bins = [-0.00072, 0.072, 0.144, 0.216, 0.288,0.360,0.432,0.504,0.576,0.648,0.72] 
num = [i for i in range(1,11)]
result2_1['总无机养分百分比标签'] = pd.cut(result2_1['总无机养分百分比'],bins,labels=num)
result2_1
```

```python
result2_1['区间'] = pd.cut(result2_1['总无机养分百分比'],10)
```

```python
result2_1
```

```python
result2_1.to_excel('result2_1.xlsx')
```

```python
pic = result2_1[['总无机养分百分比标签','区间']]
pic['num'] = 1
num = pic.groupby("总无机养分百分比标签").agg({"num":"sum"}).reset_index()
```

```python
from matplotlib import pyplot as plt
# 设置matplotlib正常显示中文和负号
plt.rcParams['font.sans-serif']=['SimHei']   # 用黑体显示中文
plt.rcParams['axes.unicode_minus']=False     # 正常显示负号

x_interval = num['总无机养分百分比标签']
width = [5,5,5,5,5,5,5,5,5,15,30,60]
y_quantity = num['num']

plt.figure(figsize=(20,8),dpi=60)

plt.bar(range(len(y_quantity)),y_quantity,width=1)

plt.grid(alpha=0.4)
plt.xlabel("总无机养分百分比标签")
plt.ylabel("num")
plt.show()
```

### 从复混肥料产品登记数量的直方图中可以看出，大部分复混肥料的总无机养分百分比都在5,6,7的这几个分组里，其他产品也都在这3个区间上下波动，证明复混肥料的总无机养分百分比占比大致为 （0.288, 0.504），其中更多的为（0.432，0.504）中
